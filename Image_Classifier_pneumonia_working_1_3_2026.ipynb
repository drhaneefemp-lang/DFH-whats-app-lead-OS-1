```python
import torch
from sklearn.metrics import confusion_matrix

model = gtf.system_dict["local"]["model"]
model.eval()

val_loader = gtf.system_dict["local"]["data_loaders"]["val"]

y_true = []
y_pred = []

with torch.no_grad():
    for inputs, labels in val_loader:
        inputs = inputs.cuda()
        outputs = model(inputs)
        _, preds = torch.max(outputs, 1)
        y_true.extend(labels.cpu().numpy())
        y_pred.extend(preds.cpu().numpy())

cm = confusion_matrix(y_true, y_pred)
print("Confusion Matrix:\n", cm)

tn, fp, fn, tp = cm.ravel()
sensitivity = tp / (tp + fn)
specificity = tn / (tn + fp)

print(f"Sensitivity: {sensitivity:.4f}")
print(f"Specificity: {specificity:.4f}")
```

```python
import torch
import numpy as np
from sklearn.metrics import confusion_matrix

model = gtf.system_dict["local"]["model"]
model.eval()

val_loader = gtf.system_dict["local"]["data_loaders"]["val"]

# Collect probabilities instead of hard predictions
y_true = []
y_probs = []

with torch.no_grad():
    for inputs, labels in val_loader:
        inputs = inputs.cuda()
        outputs = model(inputs)
        probs = torch.softmax(outputs, dim=1)
        y_probs.extend(probs[:, 1].cpu().numpy())  # probability of malignant (class 1)
        y_true.extend(labels.cpu().numpy())

y_true = np.array(y_true)
y_probs = np.array(y_probs)

# Try different thresholds
print(f"{'Threshold':<12} {'Sensitivity':<14} {'Specificity':<14} {'Accuracy':<10}")
print("-" * 50)

for threshold in [0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5]:
    y_pred = (y_probs >= threshold).astype(int)
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    sens = tp / (tp + fn)
    spec = tn / (tn + fp)
    acc = (tp + tn) / (tp + tn + fp + fn)
    print(f"{threshold:<12.2f} {sens:<14.4f} {spec:<14.4f} {acc:<10.4f}")
```

```python
import cv2, torch, numpy as np, matplotlib.pyplot as plt
from pytorch_grad_cam import GradCAM
from pytorch_grad_cam.utils.image import show_cam_on_image, preprocess_image
from pytorch_grad_cam.utils.model_targets import ClassifierOutputTarget

img_path = "/content/drive/MyDrive/pneumonia image /Test/Normal/IM-0005-0001.jpeg"
threshold = 0.07
pneumonia_idx = 1  # confirm with class_to_idx

model = gtf.system_dict["local"]["model"]
model.eval()
for p in model.parameters():
    p.requires_grad = True

img_bgr = cv2.imread(img_path, cv2.IMREAD_COLOR)
assert img_bgr is not None, f"Could not read: {img_path}"
rgb_img = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
rgb_img = cv2.resize(rgb_img, (224, 224)).astype(np.float32) / 255.0

input_tensor = preprocess_image(
    rgb_img, mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]
)

with torch.no_grad():
    logits = model(input_tensor.cuda())
    probs = torch.softmax(logits, dim=1)[0].cpu().numpy()

pneumonia_prob = float(probs[pneumonia_idx])
pred_idx = int(np.argmax(probs))
prediction = "pneumonia" if pneumonia_prob >= threshold else "normal"

target_layers = [model.features[-1]]
cam = GradCAM(model=model, target_layers=target_layers)

# Explain pneumonia score specifically (not argmax)
targets = [ClassifierOutputTarget(pneumonia_idx)]
grayscale_cam = cam(input_tensor=input_tensor.cuda(), targets=targets)[0]

visualization = show_cam_on_image(rgb_img, grayscale_cam, use_rgb=True)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1); plt.imshow(rgb_img); plt.title("Original"); plt.axis("off")
plt.subplot(1, 2, 2); plt.imshow(visualization)
plt.title(f"Grad-CAM (pneumonia target)\nPred: {prediction} | P(pneumonia)={pneumonia_prob:.4f}")
plt.axis("off")
plt.tight_layout(); plt.show()
```
